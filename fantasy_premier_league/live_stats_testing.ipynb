### API-FOOTBALL

right first things first, I will need the league id of the premier league within their system

```python
url = "https://api-football-v1.p.rapidapi.com/v2/leagues/country/england/2020"

headers = {
    'x-rapidapi-key': "58ba862574msh2d22f18a5692fefp1e9a03jsn31da5fc820dc",
    'x-rapidapi-host': "api-football-v1.p.rapidapi.com"
    }

response = requests.request("GET", url, headers=headers)

print(json.loads(response.text))
```

```python
print(json.loads(response.text)['api']['leagues'][0])
```

```python
for league in json.loads(response.text)['api']['leagues']:
    if league['name'] == "Premier League":
        print(league)
```
